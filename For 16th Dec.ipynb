```python
#this plot shows price per night on 4 different types of listings
import matplotlib.pyplot as plt

df = pd.read_csv("listing-types.csv") 


# Specify price per night and total number of reviews on 4 different types of listing  (Entire house, Hotel rooms, Private room and shared room)
X = df.iloc[2:101,[10,13]].values

plt.scatter(X[2:74,10],X[2:74,13],color='purple',marker='o',label='Entire house')
plt.scatter(X[75:80,10],X[75:80,13],color='green',marker='x',label='Hotel rooms')
plt.scatter(X[81:98,10],X[81:98,13],color='black',marker='^',label='Private room')
plt.scatter(X[99:101,10],X[99:101,13],color='yellow',marker='d',label='Shared room')

plt.xlabel('Price per night')
plt.ylabel('Total number of review')
plt.legend(loc='upper left')
plt.show()
```
